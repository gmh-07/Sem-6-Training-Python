```python
# TASK:
# write a function to add multiple parameters passed and print the addition of all of them how you will do that

def get_sum_2(*args):
    """
    Calculates and prints the sum of all numerical arguments passed to the function.

    Args:
        *args: A variable number of arguments to be summed.
        it will be the list of arguments passed to the function

    Returns:
        None
    """
    total = 0
    for num in args:
        total += num
    print(f"The sum of all parameters is: {total}")

# --- Example Usage ---

# Call the function with two parameters
get_sum_2(10, 20)

# Call the function with five parameters
get_sum_2(1, 2, 3, 4, 5)

# Call the function with no parameters
get_sum_2()

# Call the function with multiple parameters
get_sum_2(5, 10, 15, 20, 25, 30)
```

```python
# Now we are passing the key value pair in args as tuple
get_sum_2(1,2,3,op="+")
```
