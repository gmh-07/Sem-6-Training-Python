# Generators

```python
# Generators are functions that return an iterator that produces a sequence of results, one at a time, using the 'yield' keyword.
# They are memory efficient because they don't store the entire sequence in memory; instead, they generate values on the fly.
# This is particularly useful when dealing with large datasets or infinite sequences.

# Key characteristics of generators:
# 1. They are defined like regular functions but use 'yield' instead of 'return'.
# 2. When a generator function is called, it returns a generator object (an iterator) without executing the function body.
# 3. The code within the generator function only executes when 'next()' is called on the generator object or when iterated over (e.g., in a for loop).
# 4. The 'yield' keyword pauses the function's execution and saves its state. When 'next()' is called again, execution resumes from where it left off.

# Example of a simple generator:
# def my_generator():
#     yield 1
#     yield 2
#     yield 3

# # Creating a generator object
# gen = my_generator()

# # Iterating through the generator
# print(next(gen)) # Output: 1
# print(next(gen)) # Output: 2
# print(next(gen)) # Output: 3
# # print(next(gen)) # This would raise a StopIteration error

# # Generators can also be used in for loops:
# for value in my_generator():
#     print(value)

# Use cases:
# - Reading large files line by line.
# - Generating infinite sequences (e.g., Fibonacci series).
# - Implementing data streams.

# Generator expressions (similar to list comprehensions but use parentheses):
# my_gen_expr = (x * x for x in range(5))
# for num in my_gen_expr:
#     print(num)
```

```python
#yield will return some value from generator
# we can return multiple things through yield
def my_generator():
    yield(5)
    yield(10)
    yield(15)

g = my_generator()
print(g)
print(type(g))

print(next(g))
print(next(g))
print(next(g))
# print(next(g)
```

#Custom Iterator

```python
from random import random,randint,uniform



# Generate (count) number of random values between low and high

class MyIterator:
  def __init__(self,low,high,count): # constructor of the class
    self.low = low
    self.high = high
    self.count = count
    self.counter = 0

  def __iter__(self):     # Iterable: jiske upar hum iterate kar sakte hai
    return self

  def __next__(self):    # Iterable pe kya iterate karna hai ye next batata hai
    if self.counter < self.count:
      self.counter += 1
      return self.low+(self.high-self.low)*random()
    else:
      raise StopIteration

for n in MyIterator(10,25,5):
  print(n)
```

### `reduce()` Function

The `reduce()` function from the `functools` module is used to apply a particular function passed in its argument to all of the list elements mentioned in the sequence. This function is a way to perform computations on a list and return a single cumulative result.

**Syntax:**
`reduce(function, iterable[, initializer])`

**Parameters:**
- `function`: The function to be applied to the items of the iterable. This function must take two arguments.
- `iterable`: The sequence (e.g., list, tuple) to iterate over.
- `initializer` (optional): An initial value that is placed before the items of the iterable in the calculation, and serves as a default when the iterable is empty.

**How it works:**
1. `reduce()` takes the first two elements of the iterable and applies the specified `function` to them.
2. It then takes the result of that operation and the *next* element from the iterable, applying the `function` again.
3. This process continues until all elements in the iterable have been processed, yielding a single final result.

**Example from the notebook:**
In cell `ccvGRRz1pAzJ`, `reduce(lambda a,b:a+b,nums)` where `nums = [1,2,3,4,5]` works as follows:
- It takes `1` and `2`, and applies `a+b` -> `1+2 = 3`
- Then it takes `3` (the result) and `3` (the next element), applies `a+b` -> `3+3 = 6`
- Then it takes `6` (the result) and `4` (the next element), applies `a+b` -> `6+4 = 10`
- Then it takes `10` (the result) and `5` (the next element), applies `a+b` -> `10+5 = 15`

The final result is `15`.

# Decorators

```python
# Python Decorators

# A decorator is a function that takes another function as an argument and extends or changes the behavior of the latter function without explicitly modifying it.
# Decorators are often used for logging, timing execution, authentication, caching, etc.

# 1. Simple Decorator Structure
# Let's define a simple decorator that prints a message before and after a function call.

def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

# 2. Applying a decorator using the '@' syntax
# We can use the '@' symbol followed by the decorator function name directly above the function we want to decorate.

@my_decorator
def say_hello():
    print("Hello!")

say_hello()

# Common Use Cases for Decorators:
# - Logging: Log function calls, arguments, and return values.
# - Timing: Measure the execution time of functions.
# - Authentication/Authorization: Restrict access to functions based on user roles or permissions.
# - Caching: Store results of expensive function calls to avoid recomputing them.
# - Input Validation: Validate arguments before a function executes.
# - Retrying: Automatically retry a function call if it fails.
# - Unit Testing: Mocking or setting up test environments.
```

```python
def dec(func):
  def wrapper(*args,**kwargs):
    print("Before calling the decorated function")
    func(*args,**kwargs)
    print("After calling the decorated function")

  return wrapper

@dec
def print_sum(a,b):
  print(f"{a}+{b}={a+b}")

print_sum(1,2)
```

```python
#  now instead of printing we want the get_sum() function to return something

def dec(func):
  def wrapper(*args,**kwargs):
    print("Before calling the decorated function")
    result = func(*args,**kwargs)
    print("After calling the decorated function")
    return result

  return wrapper

@dec
def print_sum(a,b):
  print(f"{a}+{b}={a+b}")
  return a+b

print(print_sum(1,2))
```

```python
def loop(n):
  def dec(func):
    def wrapper(*args,**kwargs):
      print(f"Decorator initiated by loop with n={n}") # Example: using 'n' from the outer function
      print("Before calling the decorated function")
      result = func(*args,**kwargs)
      print("After calling the decorated function")
      return result
    return wrapper
  return dec

@loop(2)
def example_func(x, y):
  print(f"Executing example_func with {x}, {y}")
  return x * y

print(example_func(5, 3))
```
